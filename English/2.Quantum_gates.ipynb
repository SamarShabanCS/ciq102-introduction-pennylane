# 2.  Quantum gates

Many [quantum gates](https://docs.pennylane.ai/en/stable/introduction/operations.html#qubit-operators) are implemented in PennyLane. Some gates act on a single qubit (or wire) while others act on multiple wires. Here's a list of some commonly used gates and what they do:


* **Pauli X gate (X)** : similar to the classical NOT gate, it inverts the state of a qubit. If the qubit is in the $\vert 0 \rangle$ state, it switches to the $\vert 1 \rangle$ state and vice versa.
* **Hadamard gate (H)**: this gate transforms a qubit into a superposition of states. If the qubit is in the $\vert 0 \rangle$ state, it changes to the ($\vert 0 \rangle + \vert 1 \rangle)/\sqrt{2}$ state, and if it's in the $\vert 1 \rangle$ state, it changes to the ($\vert 0 \rangle - \vert 1 \rangle)/\sqrt{2}$ state.
* **Phase gates (S and T)**: these gates modify the phase of a qubit. The $S$ gate adds a $\pi/2$ phase to the |1‚ü© state, while the $T$ gate adds a $\pi/4$ phase.
* **Controlled-NOT gate (CNOT)**: this gate performs a $NOT$ operation on a target qubit only if a control qubit is in the $\vert 1 \rangle$ state. If the control qubit is in the $\vert 0 \rangle$ state, both qubits remain unchanged.
* **Controlled-Z gate (CZ)**: this gate adds a phase of -1 to the $\vert 11 \rangle$ state of a pair of qubits, leaving the other states unchanged.
* **Toffoli gate**: this gate performs a NOT operation on a target qubit only if two control qubits are both in the $\vert 1 \rangle$ state.
