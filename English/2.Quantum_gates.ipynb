# 2.  Quantum gates

Many [quantum gates](https://docs.pennylane.ai/en/stable/introduction/operations.html#qubit-operators) are implemented in PennyLane. There are 1 and 2 qubit gates, as well as controlled gates. Here's a list of the most common ones:


* **Quantum NOT gate (or Pauli X gate)** : similar to the classical NOT gate, it inverts the state of a qubit. If the qubit is in the |0⟩ state, it switches to the |1⟩ state and vice versa.
* **Hadamard gate (H)**: this gate transforms a qubit into a superposition of states. If the qubit is in the |0⟩ state, it changes to the (|0⟩ + |1⟩)/√2 state, and if it's in the |1⟩ state, it changes to the (|0⟩ - |1⟩)/√2 state.
* **Phase gates (S and T)**: these gates modify the phase of a qubit. The S gate adds a π/2 phase to the |1⟩ state, while the T gate adds a π/4 phase.
* **Controlled-NOT gate (CNOT)**: this gate performs a NOT operation on a target qubit only if a control qubit is in the |1⟩ state. If the control qubit is in the |0⟩ state, both qubits remain unchanged.
* **Controlled-Z gate (CZ)**: this gate adds a phase of -1 to the |11⟩ state of a pair of qubits, leaving the other states unchanged.
**Toffoli gate**: this gate performs a NOT operation on a target qubit only if two control qubits are both in the |1⟩ state.
